```python
from nrlog import NrLog
```

# 第一步：创建SV Log对象，输入参数：SV log目录， 分析时间段：格式yyyymmddhhmmss

```python
svlog = NrLog(r"D:\sv", time_interval=['20210323033500', '20210323034500'])
```

# 第二步：查看分析时间段内SV Log统计信息

```python
svlog.describle()
```

## 查看小区和UE信息

```python
svlog.get_cell_and_ue_ids()
```

# 第三步： 获取小区或者ue对象

```python
ue3 = svlog.get_ue(3)
```

# 第四步： 分析UE调度信息

```python
dl = ue3.dl
```

## 查看5s粒度的所有下行时隙的bler

```python
dl.bler_of_slot(time_bin=5).plot()
```

## 查看1s粒度的所有下行时隙bler

```python
dl.bler_of_slot(time_bin=1).plot()
```

## 查看1s粒度，反馈时隙是4的下行所有时隙的bler

```python
dl.bler_of_slot(time_bin=1,slot=4).plot()
```

## 查看调度MIMO层数信息

```python
dl.mimo_layers().plot()
```

## 查看PUCCH fmt3功率测量信息

```python
dl.pucch_pc(fmt=3).plot(figsize=(20,10))
```

## 查看1s时间粒度，UE调度总次数

```python
dl.schd_ue_cnt().plot()
```

## 查看1s时间粒度，harqfail次数信息

```python
dl.harqfail_cnt().plot()
```
